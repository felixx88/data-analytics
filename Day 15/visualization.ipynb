```python
data = [1,2,3,4,6,8,10,12]
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xlabel('No of elements', fontdict={'fontsize':15, 'color':'red'})
ax.set_ylabel('Y-value', fontdict={'fontsize':15, 'color':'orange'})
ax.plot(data)
```

```python
data = [1,2,3,4,6,8,10,12]
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xlabel('No of elements', fontdict={'fontsize':15, 'color':'red'})
ax.set_ylabel('Y-value', fontdict={'fontsize':15, 'color':'orange'})
ax.grid(True)
ax.plot(data)
```

```python
data = [1,2,3,4,6,8,10,12]
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xlabel('No of elements', fontdict={'fontsize':15, 'color':'red'})
ax.set_ylabel('Y-value', fontdict={'fontsize':15, 'color':'orange'})
ax.grid(True)
ax.set_title("Basics of Visualization", loc='right')
ax.plot(data)
```

```python
data = [1,2,3,4,6,8,10,12]
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xlabel('No of elements', fontdict={'fontsize':15, 'color':'red'})
ax.set_ylabel('Y-value', fontdict={'fontsize':15, 'color':'orange'})
ax.grid(True)
ax.set_title("Basics of Visualization", loc='right')
ax.plot(data, marker='o',markersize=10,markeredgecolor='black', markerfacecolor='blue')
```

```python
data1 = [1,2,3,4,6,8,10,12]
data2 = [11,2,4,12,15,17,18,20]
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(data1, marker='o',markersize=10,markeredgecolor='black', markerfacecolor='blue',linestyle=':',alpha=0.5)   # alpha is used to litter the color so that other line points are visible 
ax.plot(data2, marker='o',markersize=10,markeredgecolor='black', markerfacecolor='green', linestyle='--',linewidth='5')   # by default linewidth is 2 
```

```python
data = [1,2,3,4,5,6,8,10,12]
data = [12,14,16,18,20,22,24,26,28]
fig, (ax1,ax2)= plt.subplots(figsize=(10,5),ncols=2)
ax1.plot(data1, marker='o',markersize=10,markeredgecolor='black', markerfacecolor='blue',linestyle=':',alpha=0.5) 
ax2.plot(data2, marker='o',markersize=10,markeredgecolor='black', markerfacecolor='green', linestyle='--',linewidth='5')
```
